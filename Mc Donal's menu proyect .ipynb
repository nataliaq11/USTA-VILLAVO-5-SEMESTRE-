```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.3f}'.format
from math import acos, cos, sin, radians
```

```python
 #este proyecto se basará en el menú que se establece en la india en McDonald's 
#en la primera tabla podemos evidenciar sobre las diferentes categorías que tiene el menú , 
 #mostrando de este modo los tipos de hamburguesa , los gramos , calorias , grasas trans , colesterol entre otros
menu = pd.read_csv('India_Menu.csv')
menu
```

```python
# tabla 2.0 utilizamos la opción 'dummy' para agregar una columna mas en caso de querer agregar otro dato
menu['Dummy',] = 1
menu.head()
```

```python
#se puede mostrar que hay 14 columnas descriptivas y 141 productos para vender en McDonald's
#incluyendo tanto hamburguesas como condimentos y opciones
menu.shape
```

```python
#1tabla-3 describe la media en porcentajes  sobre el total de calorias , grasas saturadas , colesterol tienen
menu.describe()
```

```python
#nos muestra una tabla donde analiza que tipo de categoría es cada elemento para asi mismo elaborar las graficas
menu.info()
```

```python
#como menu items era categórica la clasificamos poniendole un numero respectivo
#a cada producto para poder volvera numerica.
menu['Menu Items'].astype('category')
```

```python
menu['Per Serve Size'] = menu['Per Serve Size'].astype('category')
```

```python
#tabla-4 nos muestra el total de categorías que nos muestra el gr del producto , 
#la cantidad de productos que no se repiten y el producto mas grande que tiene
menu.describe(include='category')
```

```python
#se modifico las categorios objetc por una categorica 
for col in ['Per Serve Size', 'Menu Items','Menu Category']:
    menu[col] = menu[col].astype('category')
    
menu.info()
```

```python
# Se puede evidenciar que el producto con mayor proteina tiene una cantidad de gr 
sns.histplot(x=menu['Protein (g)'], bins=10,  kde=True);
```

```python
# se puede evidenciar que la mayoria de productos tienen de 0-10g de grasa confirmando que Mc es perjudicial para la salud 
sns.boxplot(x=menu['Total fat (g)'],  color='lightgreen', showmeans=True);
```

```python
# se demuestra todos las categorias que tiene la base de datos 
list(menu.columns)
```

```python
# al comparar el colesterol con el tipo de menu se puede analizar que las hamburguesas gourtmet tienen un mayor
#porcentaje de colesterol con 300 mg 
ax = sns.scatterplot(x="Menu Category", y="Cholesterols (mg)", data=menu);
ax.set_title('Menu category vs Cholesterols (mg)')
```

```python
# podemos evidenciar que los menus  que tiene mayor porcentaje de energia se evalua de 1 a 800 teniendo un promedio de 250 kCal
sns.distplot( a=menu["Energy (kCal)"], hist=True, kde=False, rug=False )
```

```python
#entre mas energia porpociona el menu mas calorias genera 
plt.bar("Energy (kCal)","Total fat (g)", data = menu, color = "blue")
plt.xlabel("Energy (kCal)")
plt.ylabel("Total fat (g)")
plt.title("Energy (kCal)vsTotal fat (g)")
plt.show()
```
